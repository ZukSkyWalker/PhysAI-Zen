```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 造数据：真实函数 sin(x) + 噪声
np.random.seed(42)
X = np.linspace(0, 5, 50)
y = np.sin(X * 3) + np.random.normal(0, 0.3, X.shape)
```

```python
# NW 核回归函数
def nw_regress(x_star, X, y, h=0.3):
    weights = np.exp(- (X - x_star)**2 / (2 * h**2))
    return (weights * y).sum() / weights.sum()
```

```python
# 预测
x_grid = np.linspace(0, 5, 500)
y_pred = [nw_regress(x, X, y, h=0.3) for x in x_grid]
```

```python
plt.scatter(X, y, c='red', label='Data')
plt.plot(x_grid, np.sin(x_grid*3), 'k--', label='Original Function')
plt.plot(x_grid, y_pred, 'blue', lw=2, label='NW prediction (h=0.3)')
plt.legend()
plt.title("Nadaraya-Watson Kernal Regression")
plt.show()
```
